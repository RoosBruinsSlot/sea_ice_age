```python
import glob
from datetime import datetime
import os
from multiprocessing import Pool

import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt

from uncertainty import ComputeSidSicUncertainty
%matplotlib inline
```

```python
n_steps = 2210 # should be 2210 to cover 6 years

sia_dir = r"D:\SIA_Weddell_Sea"
mesh_dir = f'{sia_dir}/mesh'
unc_dir = f'{sia_dir}/unc'
sid_dir =  r"D:\OSI-SAF\Preprocessing_Antarctica"
ifiles = sorted(glob.glob(f'{sid_dir}/20*/ice_drift_sh_ease2*.nc.npz'))
idates = [datetime.strptime(os.path.basename(ifile).split('-')[-1].split('.')[0], '%Y%m%d%H%M%S')
          for ifile in ifiles]
print(f'Found {len(ifiles)} ice drift files from {idates[0]} to {idates[-1]}')
```

```python
start_index = idates.index(datetime(2015, 3, 1, 12))

comp = ComputeSidSicUncertainty(sia_dir)
comp.save_sid_sic_uncertainty(idates[start_index])
```
